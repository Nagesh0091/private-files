### üî§ **LABEL ENCODING**

üéØ **Purpose**: Convert **categories into numbers**.

üìå **Best For**: **Ordinal data** (where order matters).

‚ÄÉ‚ÄÉ‚û°Ô∏è e.g., 'Low' < 'Medium' < 'High'

üß† **How It Works**:
Each unique category gets a **unique integer** value.



````python
from sklearn.preprocessing import LabelEncoder

LE=LabelEncoder()

DATA["event_type_LE"]=LE.fit_transform(DATA["event_type"])
````


### üî≤ **ONE-HOT ENCODING**

üéØ **Purpose**: Represent **categories as binary columns**.

üìå **Best For**: **Nominal data** (no natural order).

‚ÄÉ‚ÄÉ‚û°Ô∏è e.g., 'Red', 'Green', 'Blue' have no ranking.

üß† **How It Works**:
Creates a **new column for each category** with `1` or `0`.


‚úÖ **Pro**: No assumption of order.
‚ö†Ô∏è **Con**: Can create **many columns** if there are many categories.

### üßæ **TRANSACTION ENCODER**

üéØ **Purpose**: Convert **lists of items** (like shopping carts) into binary format.

üìå **Best For**: **Market Basket data** or **association rule mining**.

‚ÄÉ‚ÄÉ‚û°Ô∏è e.g., \[\['milk', 'bread'], \['bread', 'butter']]

üß† **How It Works**:
Each row is a transaction. Each column represents an item ‚Üí `1` if present, `0` if not.

üìä **Example**:

```
Input:
[['milk', 'bread'],
 ['bread', 'butter']]

Encoded:
bread  butter  milk  
  1       0      1  
  1       1      0  
```

‚úÖ **Pro**: Ideal for **Apriori** or **FP-Growth** algorithms.
‚ö†Ô∏è **Con**: Works only with **lists of items** (not single columns).

---

### üî¢ **ORDINAL ENCODING**

üéØ **Purpose**: Convert **categories with a natural order** into integers.

üìå **Best For**: **Ordered data** (e.g., low, medium, high).

‚ÄÉ‚ÄÉ‚û°Ô∏è e.g., 'Small' < 'Medium' < 'Large'

üß† **How It Works**:
Each category gets a **unique integer** based on its **rank**.

üìä **Example**:

```
Input:     ['Small', 'Medium', 'Large']  
Encoded:   [   1 ,      2  ,    3   ]
```

‚úÖ **Pro**: Maintains the **order** of categories.
‚ö†Ô∏è **Con**: Not suitable for **nominal data** (no order).

---


````python
# Sample data
DATA_OE = [['Low'], ['Medium'], ['High'], ['Medium'], ['Low']]
DATA_OE=pd.DataFrame(DATA_OE,columns=["temp"])
DATA_OE
````



### üëú **Bag of Words (BoW)**

üéØ **Purpose**: Capture **word frequency** in text.

üìå **How It Works**: Converts text into a matrix where each column represents a unique word, and values are the frequency of each word in the document.

‚úÖ **Pro**: Simple and interpretable.

‚ö†Ô∏è **Con**: Ignores **word order** and context.

---





````python
# BAG OF WORD :
from sklearn.feature_extraction.text import CountVectorizer

CV=CountVectorizer()
BAG_OF_WORDS=CV.fit_transform(DATA["Document"])
BAG_OF_WORDS=pd.DataFrame(BAG_OF_WORDS.toarray(),columns=CV.get_feature_names_out())
BAG_OF_WORDS
````

### üß† **TF-IDF (Term Frequency-Inverse Document Frequency)**

üéØ **Purpose**: Highlight **important words** based on both frequency and uniqueness.

üìå **How It Works**: Weighs words by how often they appear in a document (TF) and how rare they are across all documents (IDF).

‚úÖ **Pro**: Focuses on **rare and informative words**.

‚ö†Ô∏è **Con**: Computationally **intensive**.

---

````python
# TF-IDF:
from sklearn.feature_extraction.text import TfidfVectorizer
TF=TfidfVectorizer()

TFIDF=TF.fit_transform(DATA["Document"])
TFIDF=pd.DataFrame(TFIDF.toarray(),columns=TF.get_feature_names_out())
TFIDF
````

### üåê **Word Embeddings**

üéØ **Purpose**: Capture deeper **semantic meanings** of words.

üìå **How It Works**: Words are represented as dense vectors in continuous space using models like **Word2Vec**, **GloVe**, or **FastText**.

‚úÖ **Pro**: Captures **word relationships** (e.g., "king" - "man" + "woman" ‚âà "queen").

‚ö†Ô∏è **Con**: Requires **large amounts of data** to train effectively.

---



````python
tokens = [text.lower().split() for text in DATA["Document"]]
tokens

````

````python
from gensim.models import Word2Vec
W2V=Word2Vec(sentences=tokens, vector_size=50, window=2, min_count=1)
W2V
````


### üß≥ **Pretrained Embeddings (e.g., BERT)**

üéØ **Purpose**: Provide **contextualized word representations**.

üìå **How It Works**: Leverages **advanced models like BERT** that consider the context in which a word is used, providing rich, **context-dependent embeddings**.

‚úÖ **Pro**: **State-of-the-art** performance in many NLP tasks.

‚ö†Ô∏è **Con**: **Resource-intensive** (large model sizes, expensive to compute).

````python
from transformers import BertTokenizer, BertModel
import torch

tokenizer = BertTokenizer.from_pretrained("bert-base-uncased")
model = BertModel.from_pretrained("bert-base-uncased")

input_ids = tokenizer("I love data science", return_tensors="pt")
outputs = model(**input_ids)

# Last hidden state for each token
print(outputs.last_hidden_state.shape)  # (1, sequence_length, 768)
````

#### **All these methods convert text into numerical features in unique ways, which directly affect how well the model performs.**

* **Bag of Words** and **TF-IDF** focus on **word frequency**,

* **Word Embeddings** and **Pretrained Models** capture **deeper word meanings** and **context**.

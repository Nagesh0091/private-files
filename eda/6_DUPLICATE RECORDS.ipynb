üéØ **Effect of Duplicates on Mean** üéØ  
> **"The mean is often seen as the heart of the dataset."**

üìå It **summarizes the central tendency**, giving us a single value to represent the entire data.  
BUT... it **fails** when the data is **unfair or extreme**!

üö´ **When does the mean mislead us?**  
- ‚ö†Ô∏è **Outliers**: A few extreme values can drag the mean too high or too low.  
- üìâ **Skewed Data**: If the data leans heavily to one side, the mean gets pulled away from the true center.  
- üîÅ **Duplicates**: Repeating values ‚Äî especially extreme ones ‚Äî can **magnify** the distortion!

‚úÖ **There‚Äôs only one solution:**  
üßπ **Remove duplicate records** to ensure the mean truly represents your data!

## 1. **One-Sample t-Test** ğŸ§‘â€ğŸ«

**Used when:**  
You have a **single sample** and want to compare its **mean** to a known **population mean**.

**Example:**  
You have a sample of 25 students' test scores, and you want to know if the **average score** differs from the population mean of 75.

### Hypothesis:
- **Null Hypothesis (Hâ‚€):** The sample mean is equal to 75.  
  \[
  H_0: mu = 75
  \]
- **Alternative Hypothesis (Hâ‚):** The sample mean is not equal to 75.  
  \[
  H_1: mu != 75
  \]

---



```python
# ONE SAMPLE T-TEST
marks = [11,17,18,14,19,19.5,18,16,13,17,16,14,15,18,19,11,10,7,2,9,3,14,17,16,19]
population_mean = 75

# Hypotheses
ho = "sample mean = population mean"
ha = "sample mean != population mean"

# Perform one-sample t-test
from scipy import stats
t_statistic, p_value = stats.ttest_1samp(marks, population_mean)

print("t-statistic:", t_statistic)
print("p-value:", p_value)

# Significance level
ALPHA = 0.05

# Decision
if p_value < ALPHA:
    print(f"Reject null hypothesis. {ha}")
else:
    print(f"Fail to reject null hypothesis. {ho}")
```

## 2. **Two-Sample t-Test** ğŸ†š

**Used when:**  
You want to compare the **means** of **two independent samples** to see if they are **significantly different**.

**Example:**  
You want to compare the average height of male and female students in a class.

### Hypothesis:
- **Null Hypothesis (Hâ‚€):** The mean height of males is equal to the mean height of females.  
  [
  H_0: mu_1 = mu_2
  ]
- **Alternative Hypothesis (Hâ‚):** The mean height of males is not equal to the mean height of females.  
  [
  H_1: mu_1 != mu_2
  ]

---



```python
# TWO SAMPLE T-TEST:
# TWO SAMPLES ARE GIVE SAID TO COMPARE BOTH SAMPLES MEAN

MALE_FOOTS=[4,5,6,7,4,5,6,7,5,6,7,5.8,6.3,5.7,5.5,5.3,4.9]
FEMALE_FOOTS=[3.8,4.2,4.8,4.9,4.5,5.2,5.1,4,5.4,4.6,4,4.9,5.5,5,5.3,5.9,6]
MU1=sum(MALE_FOOTS)/len(MALE_FOOTS)
MU2=sum(FEMALE_FOOTS)/len(FEMALE_FOOTS)

HO="MALE HEIGHT MEAN IS EQUAL TO FEMALE HIEGHT MEAN"
HA="MALE HEIGHT MEAN IS NOT EQUAL TO FEMALE HEIGHT MEAN"

from scipy import stats

t_statistic, p_value = stats.ttest_ind(MALE_FOOTS, FEMALE_FOOTS, equal_var=False)  # assume equal variances

print("t-statistic:", t_statistic)
print("p-value:", p_value)


# COMPARISON OF P_VALUE AND ALPHA:
ALPHA=0.05

if p_value<ALPHA:
    print(f"reject null hypothesis {HA}")
else:
    print(f"accept null hypothesis {HO}")    
```

## 3. **One-Sample One-Tailed t-Test** (Right or Left) â¡ï¸ â¬…ï¸

**Used when:**  
You test if the sample mean is either **greater than** or **less than** the population mean. It tests only in **one direction**.

### Right-Tailed Test (Greater Than):
**Example:**  
You want to test if the average temperature in a city is **greater** than 30Â°C.

### Hypothesis:
- **Null Hypothesis (Hâ‚€):** The mean temperature is less than or equal to 30Â°C.  
  [
  H_0: mu < 30
  ]
- **Alternative Hypothesis (Hâ‚):** The mean temperature is **greater** than 30Â°C.  
  [
  H_1: mu > 30
  ]

### Left-Tailed Test (Less Than):
**Example:**  
You want to test if the average temperature in a city is **less** than 30Â°C.

### Hypothesis:
- **Null Hypothesis (Hâ‚€):** The mean temperature is greater than or equal to 30Â°C.  
  [
  H_0: mu > 30
  ]
- **Alternative Hypothesis (Hâ‚):** The mean temperature is **less** than 30Â°C.  
  [
  H_1: mu < 30
  ]

---



```python
# ONE SAMPLE ONE TAILED TEST:
# Given data:
temperatures = [29.5, 30.1, 30.3, 30.8, 31.2, 30.5, 29.9, 31.5, 30.7, 30.9,
                31.0, 30.6, 30.2, 30.3, 31.4, 31.6, 30.8, 30.9, 31.1, 30.4]

# Population mean to compare with:
population_mean = 30

# stats the hypothesis:
ho="sample mean > 30"
ha="sample mean< 30"

# Perform one-sample TWO TAILED t-test... 
from scipy import stats
t_stat, p_value_two_tailed = stats.ttest_1samp(temperatures, population_mean)

# CONVERT P_VALUE_TWO_TAILED TO ONE TAIL P_VALUE:
p_value=p_value_two_tailed/2


# Print results:
import numpy as np
print("population_mean :",population_mean)
print("Sample Mean:", np.mean(temperatures))
print("t-statistic:", t_stat)
print("p-value (one-tailed):", p_value)

# Significance level:
alpha = 0.05   

# Decision:
if p_value < alpha:
    print("Reject Hâ‚€{ha} â†’ The average temperature is significantly greater than 30Â°C.")
else:
    print("accept null hypothesis{ho} â†’ Not enough evidence that the average temperature is greater than 30Â°C.")

```

## 4. **Two-Sample One-Tailed t-Test** (Right or Left) â¡ï¸ â¬…ï¸

**Used when:**  
You want to compare **two independent sample means** in a **specific direction** (greater or lesser).

### Right-Tailed Test (Greater Than):
**Example:**  
You want to test if the **average salary** of employees in Company A is **greater** than in Company B.

### Hypothesis:
- **Null Hypothesis (Hâ‚€):** Company A salary is **less than or equal to** Company B salary.  
  [
  H_0: mu_1 < mu_2
  ]
- **Alternative Hypothesis (Hâ‚):** Company A salary is **greater** than Company B salary.  
  [
  H_1: mu_1 > mu_2
  ]

### Left-Tailed Test (Less Than):
**Example:**  
You want to test if the **average salary** of employees in Company A is **less** than in Company B.

### Hypothesis:
- **Null Hypothesis (Hâ‚€):** Company A salary is **greater than or equal to** Company B salary.  
  [
  H_0: mu_1 >mu_2
  ]
- **Alternative Hypothesis (Hâ‚):** Company A salary is **less** than Company B salary.  
  [
  H_1: mu_1 < mu_2
  ]

---



```python
company_a = [55, 60, 58, 62, 65, 59, 61, 64, 66, 63]
company_b = [52, 50, 49, 55, 53, 54, 51, 50, 52, 54]
import numpy as np
co_a_mean=np.mean(company_a)
co_b_mean=np.mean(company_b)

ho="Company A salary is less than or equal to Company B salary."
ha="Company A salary is greater than Company B salary"

# Perform two-sample one tailed t-test...
t_stat, p_two_tailed = stats.ttest_ind(company_a, company_b, equal_var=True)

# CONVERT P_VALUE_TWO_TAILED TO ONE TAIL P_VALUE:
p_value=p_two_tailed/2

# Print results
print("Company A Mean:", co_a_mean)
print("Company B Mean:", co_b_mean)
print("t-statistic:", t_stat)
print("p-value (one-tailed):", p_value)


# Significance level
alpha = 0.05

if p_value<alpha:
    print(f"reject null hypothesis:{ha}")
else:
    print(f"accept null hypothesis{ho}")    

```

## 5. **One-Sample Two-Tailed t-Test** ğŸ”„

**Used when:**  
You want to test if the sample mean is **different** from the population mean in **both directions** (greater or less).

**Example:**  
You want to test if the average number of visitors to a website is **different** from 1,000.

### Hypothesis:
- **Null Hypothesis (Hâ‚€):** The mean number of visitors is 1,000.  
  [
  H_0: mu = 1000
  ]
- **Alternative Hypothesis (Hâ‚):** The mean number of visitors is **not equal** to 1,000.  
  [
  H_1: mu != 1000
  ]

---



```python
# ONE SAMPLE TWO TAILED TEST:

# Sample data: daily visitors to the website over 15 days
visitors = [980, 1020, 995, 1010, 990, 985, 1005, 1015, 970, 1000, 980, 1005, 995, 1010, 990]

# Population mean (assumed benchmark)
population_mean = 1000

# Hypotheses
ho = "Mean number of visitors = 1000"
ha = "Mean number of visitors != 1000"

# Perform one-sample t-test :
from scipy import stats
t_stat, p_value = stats.ttest_1samp(visitors,population_mean)
# P_VALUE IS TWO TAILED ALREADY...

# Significance level
alpha = 0.05

# Results
import numpy as np
print("Sample Mean:", np.mean(visitors))
print("t-statistic:", t_stat)
print("p-value:", p_value)

# Conclusion
if p_value < alpha:
    print("REJECT NULL HYPOTHESISHâ‚€:", ha)
else:
    print("ACCEPT NULL HYPOTHESIS:", ho)
```

## 6. **Two-Sample Two-Tailed t-Test** ğŸ”„ğŸ”„

**Used when:**  
You are comparing the **means** of **two independent samples** and testing if the means are **different** (either greater or less).

**Example:**  
You want to compare the **test scores** of students from two different schools and see if they differ.

### Hypothesis:
- **Null Hypothesis (Hâ‚€):** The average scores of the two schools are equal.  
  [
  H_0: mu_1 = mu_2
  ]
- **Alternative Hypothesis (Hâ‚):** The average scores of the two schools are **not equal**.  
  [
  H_1: mu_1 != mu_2
  ]

---


```python
# TWO SAMPLE TWO TAILED TEST:
# Test scores
school_a = [78, 85, 92, 88, 76, 81, 90, 84, 79, 87]
school_b = [72, 80, 78, 74, 77, 73, 76, 75, 79, 78]

import numpy as np
school_a_mean=np.mean(school_a)
school_b_mean=np.mean(school_b)

ho="school_a_mean = school_b_mean"
ha="school_a_mean != school_b_mean"

# perform TWO SAMPLE TWO TAILED TEST:
t_stat,p_value=stats.ttest_ind(school_a,school_b,equal_var=True)

# print details:
print(f"mean of school a test scores {school_a_mean}")
print(f"mean of school b test scores {school_b_mean}")
print(f"t_statistics : {t_stat}")
print(f"p_value(two tailed): {p_value}")

# Significance level
alpha = 0.05

# conclusion:
if p_value<alpha:
    print(f"reject null hypothesis {ha}")
else:
    print(f"accept null hypithesis {ho}")
```

# 7. **Two-Proportion t-Test** ğŸ“Š

### **Used When:**
You want to compare the **proportions** of two independent groups to see if they are **significantly different**.

**Example:**  
You want to compare the proportion of **male and female students** who passed an exam in two different classes.

### **Hypothesis:**
- **Null Hypothesis (Hâ‚€):** The proportion of male students who passed is equal to the proportion of female students who passed.  
  [
  H_0: p_1 = p_2
  ]
- **Alternative Hypothesis (Hâ‚):** The proportion of male students who passed is not equal to the proportion of female students who passed.  
  [
  H_1: p_1 != p_2
  ]

---





### âœ… **Data**  
- Males: 60 students, 48 passed  
- Females: 70 students, 49 passed  


```python
# TWO PROPORTION T-TEST:
# Data
success = [48, 49]        # number of students who passed
Tnos = [60, 70]           # total number of students


# Hypotheses
ho = "Proportion of males who passed = proportion of females who passed"
ha = "Proportion of males who passed â‰  proportion of females who passed"


# Perform two-proportion Z-test
from statsmodels.stats.proportion import proportions_ztest
z_stat, p_value = proportions_ztest(success, Tnos)

# Significance level
alpha = 0.05


# Results
print("Z-statistic:", z_stat)
print("p-value:", p_value)

# Conclusion
if p_value < alpha:
    print("Reject Hâ‚€:", ha)
else:
    print("Fail to reject Hâ‚€:", ho)
```

# 10. **ANOVA Test (Analysis of Variance)** ğŸ“Š

### **Used When:**
You want to compare the **means** of **three or more groups** to see if there is a significant difference in their means.

**Example:**  
You want to compare the **average test scores** of students in three different study programs (A, B, and C) to see if the means differ.

### **Hypothesis:**
- **Null Hypothesis (Hâ‚€):** The mean test scores for all study programs are equal.  
  [
  H_0: mu_1 = mu_2 = mu_3
  ]
- **Alternative Hypothesis (Hâ‚):** At least one of the mean test scores is different.  
  [
  H_1: At least one mu  is different.
  ]

**Note:**  
ANOVA doesn't tell you which groups are differentâ€”just that there is a difference. If **Hâ‚** is true, post-hoc tests (e.g., Tukey's HSD) are often conducted to determine which groups differ.

---



```python
# ANOVA ANALYSIS OF VARIENCE:
# Scores from 3 different study programs
program_A = [85, 88, 90, 92, 87]
program_B = [78, 74, 80, 79, 77]
program_C = [90, 92, 88, 91, 89]

# Hypotheses
ho = "The mean test scores for all study programs are equal."
ha = "At least one of the mean test scores is different."

# Perform one-way ANOVA
from scipy import stats
f_stat, p_value = stats.f_oneway(program_A, program_B, program_C)

# Significance level
alpha = 0.05

# Results
print("F-statistic:", f_stat)
print("p-value:", p_value)

# Conclusion
if p_value < alpha:
    print("Reject Hâ‚€:", ha)
else:
    print("Fail to reject Hâ‚€:", ho)
```

```python
# Combine all scores and group labels:
scores = [85, 88, 90, 92, 87, 78, 74, 80, 79, 77, 90, 92, 88, 91, 89]
groups = ['A']*5 + ['B']*5 + ['C']*5

# Create DataFrame
import pandas as pd
df = pd.DataFrame({'score': scores, 'group': groups})

# Tukey HSD test
from statsmodels.stats.multicomp import pairwise_tukeyhsd
tukey_result = pairwise_tukeyhsd(endog=df['score'], groups=df['group'], alpha=0.05)
print(tukey_result)
```

# 11. **Chi-Square Test** ğŸ”¢

### **Used When:**
You want to test the **association** between two **categorical variables** or compare the **observed** and **expected frequencies** in a contingency table.

There are two main types of Chi-Square tests:
1. **Chi-Square Goodness of Fit Test:** Tests if an observed distribution matches a specific distribution.
2. **Chi-Square Test of Independence:** Tests if two categorical variables are independent.

### **Example 1:** **Chi-Square Goodness of Fit**  
You want to test if the distribution of **coin flips** (heads or tails) is **fair** (50% heads, 50% tails).

### **Hypothesis:**
- **Null Hypothesis (Hâ‚€):** The coin is fair (heads and tails are equally likely).  
  [
  H_0: heads = tails = 0.5
  ]
- **Alternative Hypothesis (Hâ‚):** The coin is **not fair** (the probabilities are not equal).  
  [
  H_1: heads != tails != 0.5
  ]



```python
'''
ğŸ”¹ 1. Chi-Square Goodness of Fit Test
You compare observed frequencies with expected frequencies.

ğŸª™ Example: Fair Coin Toss
Suppose you flip a coin 100 times: 55 heads, 45 tails.
'''
ho = "head and tail frequencies are equal"
ha = "head and tail frequency are not equal"

# Observed frequencies
observed = [55, 45]

# Expected frequencies if fair coin
expected = [50, 50]

# Perform Chi-Square Goodness of Fit Test
from scipy.stats import chisquare
chi2_stat, p_val = chisquare(f_obs=observed, f_exp=expected)

print("Chi-square Statistic:", chi2_stat)
print("p-value:", p_val)

alpha = 0.05

# compare p_val with alpha:
if p_val < alpha:
    print(f"Reject null hypothesis: {ha}")
else:
    print(f"Accept null hypothesis: {ho}")
    
'''
âœ… Output Meaning:
If p_val < 0.05: Significant result â†’ coin might not be fair.

If p_val â‰¥ 0.05: No significant evidence â†’ coin is likely fair.'''        
```

### **Example 2:** **Chi-Square Test of Independence**  
You want to test if there is an association between **gender** (male/female) and **job satisfaction** (satisfied/unsatisfied) in a company.

### **Hypothesis:**
- **Null Hypothesis (Hâ‚€):** There is no association between gender and job satisfaction (they are independent).  
  [
  H_0:Gender and Job Satisfaction are independent.
  ]
- **Alternative Hypothesis (Hâ‚):** There is an association between gender and job satisfaction (they are not independent).  
  [
  H_1:Gender and Job Satisfaction are dependent.
  ]

---

```python
# Hypotheses
ho = "Gender and Job Satisfaction are independent"
ha = "Gender and Job Satisfaction are dependent"

# Contingency Table:
#                Satisfied  Unsatisfied
#    Male            40          10
#    Female          30          20


data = np.array([[40, 10],
                 [30, 20]])

# Perform Chi-Square Test of Independence
from scipy.stats import chi2_contingency
chi2_stat, p_val, dof, expected = chi2_contingency(data)# double dimension numpy array


print("Chi-square Statistic:", chi2_stat)
print("p-value:", p_val)
print("Degrees of Freedom:", dof)
print("Expected Frequencies:\n", expected)


# Significance Level
alpha = 0.05

# Decision
if p_val < alpha:
    print(f"Reject null hypothesis: {ha}")
else:
    print(f"Accept null hypothesis: {ho}")


```

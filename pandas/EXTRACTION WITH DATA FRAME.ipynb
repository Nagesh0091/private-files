# üìö `.iloc` Selection

---

### ‚ú® Extracting all Rows and single column
```python
df.iloc[:, 1]
```

---

### ‚ú® Extracting all Rows and range of columns only
```python
df.iloc[:, 1:3]
```

---

### ‚ú® Extracting all Rows and single column
```python
df.iloc[:, 1]
```

---

### ‚ú® Extracting top 5 Rows and range of columns only
```python
df.iloc[0:5, 1:4]
```

---

### ‚ú® Extracting top 5 Rows and single column
```python
df.iloc[0:5, 1]
```

---

### ‚ú® Extracting all columns and single Row
```python
df.iloc[2, :]
```
or
```python
df.iloc[[2]]
```

---

### ‚ú® Extracting random records by using index parameter
```python
df.iloc[[0, 2, 4]]
```
or
```python
df.iloc[[0, 2, 4], :]
```

---

### ‚ú® Skipped Rows (start:stop:endvalue/stepvalue)
```python

```

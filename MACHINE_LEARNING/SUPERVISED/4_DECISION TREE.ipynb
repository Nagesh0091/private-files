### Decision Tree Algorithm üå≥

A **Decision Tree** is a model that splits data into branches based on certain conditions to make predictions. It looks like a tree with a root (starting point) and leaves (final predictions). Each node represents a decision based on a feature, and branches are formed based on that decision. üå±

---

### Key Details to Know üåü

1. **Structure** üå≥:

   * **Root**: The starting point of the tree.
   * **Nodes**: Represent decisions made based on feature values.
   * **Leaves**: The final prediction or outcome.
   * **Branches**: The path between nodes.

2. **Working** üîç:

   * The tree splits data by testing a feature at each node.
   * It selects the best feature (based on criteria like **Gini Impurity** or **Information Gain**) to split the data.
   * The process continues until the data is classified or a stopping condition is met (like max depth or min samples per leaf).

3. **Advantages** üéâ:

   * Easy to understand and visualize.
   * Handles both numerical and categorical data.
   * Non-linear relationships can be captured.
   * Little data preparation needed (no scaling required).

4. **Disadvantages** ‚ö†Ô∏è:

   * Prone to **overfitting** (learns noise from data).
   * Unstable (small changes in data can lead to different trees).
   * Can be biased toward features with more levels.

5. **Use Cases** üíº:

   * Classification (e.g., spam detection üì®, customer churn prediction üíî).
   * Regression (e.g., predicting house prices üè†).

6. **Important Parameters** üîß:

   * **Max Depth**: Controls how deep the tree can grow.
   * **Min Samples Split**: Minimum samples required to split a node.
   * **Criterion**: How to measure the quality of a split (e.g., Gini Impurity, Entropy).

7. **Avoiding Overfitting** üõ°Ô∏è:

   * **Pruning**: Cutting off branches that provide little value.
   * **Cross-Validation**: To test the model on different data sets.



# Types of Decision Tree Algorithms üå≥ (With C5.0 First)

1. **C5.0 üí°**

   * **How it works**: An advanced version of C4.5 that uses **boosting** to improve model accuracy by combining weak learners into a stronger one. It is faster and more efficient than C4.5.
   * **Difference**:

     * Uses **boosting** and **entropy** for splits.
     * **Post-pruning** to avoid overfitting.
     * **Smaller trees** for better performance and efficiency.
   * **Best for**: Large-scale **classification** tasks with high accuracy.
   * **Limitation**: Not as widely supported as C4.5 or CART in some tools.

2. **CART (Classification and Regression Tree) üìä**

   * **How it works**: Uses **Gini Impurity** for classification and **Mean Squared Error** for regression. It builds **binary trees** (each node has two branches).
   * **Difference**: Can handle both **classification** and **regression**. It is known for its simplicity and versatility.
   * **Best for**: Both **classification** and **regression** tasks.
   * **Limitation**: Can be prone to overfitting if not properly pruned.

3. **ID3 (Iterative Dichotomiser 3) üåê**

   * **How it works**: Uses **Entropy** and **Information Gain** to decide the best feature for splitting the data.
   * **Difference**: It maximizes **Information Gain** to reduce entropy.
   * **Best for**: Simple **classification** tasks.
   * **Limitation**: Prone to overfitting and struggles with continuous data.

4. **C4.5 üìö**

   * **How it works**: An improvement over ID3 that uses **Gain Ratio** instead of Information Gain and handles continuous data.
   * **Difference**: Has **pruning** to avoid overfitting and can handle both **categorical and continuous features**.
   * **Best for**: Complex **classification** tasks.
   * **Limitation**: More computationally expensive than ID3.

5. **CHAID (Chi-squared Automatic Interaction Detection) üßë‚Äçüî¨**

   * **How it works**: Uses the **Chi-squared test** to find the best feature to split data.
   * **Difference**: Focuses on statistical significance for splitting.
   * **Best for**: Data with categorical variables.
   * **Limitation**: Assumes a large sample size for meaningful results.

6. **M5 Algorithm üíª**

   * **How it works**: Combines decision trees with **linear regression** models at the leaves for regression tasks.
   * **Difference**: Merges decision tree structure with **linear regression** to model continuous data.
   * **Best for**: **Regression** problems with complex relationships.
   * **Limitation**: Less effective for classification.

---

### Key Differences Between Them üîë

* **Split Criteria**:
  
  * **C5.0**: Boosting + Entropy.
  * **CART**: Gini Impurity (classification), MSE (regression),
  * **ID3**: Information Gain,
  * **C4.5**: Gain Ratio,
  * **CHAID**: Chi-squared test,
  

* **Pruning**:

  * **C4.5**, **C5.0**, and **CART** can prune trees, whereas **ID3** doesn't have pruning by default.

* **Binary vs Multi-way Splits**:

  * **CART** builds **binary trees**, while **ID3**, **C4.5**, and **C5.0** can handle multi-way splits.

* **Regression Support**:

  * **CART** and **M5** can handle **regression**, while others are mostly used for **classification**.

* **Boosting**:

  * **C5.0** is unique in using **boosting** to improve accuracy by combining multiple trees.

```python

# DATA LOADING :
import pandas as pd
DATA = pd.read_csv(r"C:\Users\Nagesh Agrawal\OneDrive\Desktop\MACHINE LEARNING\1_DATASETS\DECISION TREE DATA.csv")

# Label Encoding
from sklearn.preprocessing import LabelEncoder
LE = LabelEncoder()
DATA["Species"] = LE.fit_transform(DATA["Species"])

# Selecting Features and Target
X = DATA.iloc[:, 1:5]   # Features
Y = DATA["Species"]     # Target

# Splitting the data
from sklearn.model_selection import train_test_split
X_TRAIN, X_TEST, Y_TRAIN, Y_TEST = train_test_split(X, Y, test_size=0.3, random_state=2)

# DECISION TREE USING ENTROPY (C5.0-like)
from sklearn.tree import DecisionTreeClassifier
DTC = DecisionTreeClassifier(criterion="entropy", max_depth=2, random_state=2)
DTC.fit(X_TRAIN, Y_TRAIN)

import matplotlib.pyplot as plt
from sklearn import tree

plt.figure(figsize=(12, 8))
FEATURE_NAME = DATA.columns[:-1].tolist()
# Convert class names to list of strings
CLASS_NAME = DATA["Species"].unique().astype(str).tolist()
# Plot the tree
tree.plot_tree(DTC, feature_names=FEATURE_NAME, class_names=CLASS_NAME, filled=True)
plt.show()

# üîç Make predictions
Y_PRED = DTC.predict(X_TEST)

# üìä Check accuracy
from sklearn.metrics import accuracy_score
accuracy = accuracy_score(Y_TEST, Y_PRED)
print("Accuracy:", accuracy)
```

```python

# DATA LOADING :
import pandas as pd
DATA = pd.read_csv(r"C:\Users\Nagesh Agrawal\OneDrive\Desktop\MACHINE LEARNING\1_DATASETS\DECISION TREE DATA.csv")

# Label Encoding
from sklearn.preprocessing import LabelEncoder
LE = LabelEncoder()
DATA["Species"] = LE.fit_transform(DATA["Species"])

# Selecting Features and Target
X = DATA.iloc[:, 1:5]   # Features
Y = DATA["Species"]     # Target

# Splitting the data
from sklearn.model_selection import train_test_split
X_TRAIN, X_TEST, Y_TRAIN, Y_TEST = train_test_split(X, Y, test_size=0.3, random_state=2)

# CART DECISION TREE USING GINI
from sklearn.tree import DecisionTreeClassifier

# Create the CART model
CART_DTC = DecisionTreeClassifier(criterion="gini", max_depth=2, random_state=2)
CART_DTC.fit(X_TRAIN, Y_TRAIN)

# Plot the CART decision tree
import matplotlib.pyplot as plt
from sklearn import tree

plt.figure(figsize=(12, 8))
FEATURE_NAME = DATA.columns[:-1].tolist()
CLASS_NAME = DATA["Species"].unique().astype(str).tolist()

tree.plot_tree(CART_DTC, feature_names=FEATURE_NAME, class_names=CLASS_NAME, filled=True)
plt.show()

# üîç Make predictions
Y_PRED = CART_DTC.predict(X_TEST)

# üìä Check accuracy
from sklearn.metrics import accuracy_score
accuracy = accuracy_score(Y_TEST, Y_PRED)
print("Accuracy:", accuracy)
```

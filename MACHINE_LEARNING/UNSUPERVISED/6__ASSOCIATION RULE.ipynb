## 📌 **Association Rule Algorithms** –

### 🔍 **Definition (In Simple Words):**

Association rule algorithms find relationships between items in large datasets.

💡 *“If a person buys bread 🍞, they are likely to buy butter 🧈 too.”*

---

### 📖 **Key Terms to Know:**

1. **🛒 Itemset:**

   A group of items bought together (e.g., {milk, bread}).

2. **📏 Support:**

   How often an itemset appears in the data.

   🧮 *Support(A) = (Transactions with A) / (Total transactions)*

3. **📈 Confidence:**

   How often item B is bought when item A is bought.

   🧮 *Confidence(A → B) = Support(A and B) / Support(A)*

4. **📊 Lift:**

   Measures how much more likely B is bought when A is bought, compared to chance.

   🧮 *Lift = Confidence / Support(B)*

   ✅ Lift > 1 → Positive association



### ⚙️ **Popular Algorithms:**

1. **Apriori 🧮:**

   Checks all possible combinations of items using a bottom-up approach.

   🚧 Can be slow if dataset is large.

2. **Eclat 🚀:**

   Uses a depth-first search with set intersections.

   ✅ Faster than Apriori on some datasets.

3. **FP-Growth 🌲:**

   Uses a compact tree structure (FP-tree) to find frequent patterns without generating all combinations.

   ⚡ Very fast and memory efficient.



### 📂 **Where It's Used:**

* 🛍️ Market Basket Analysis

* 🎥 Recommendation Systems

* 📊 Customer Behavior Analysis

* 💊 Medical Diagnosis (symptom → disease)



### 💡 **Example Rule:**

If {diapers, milk} → {baby wipes}

* Support = 15%
* Confidence = 70%
* Lift = 2.3

✅ Means baby wipes are 2.3x more likely to be bought with diapers and milk.
